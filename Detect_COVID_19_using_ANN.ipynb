```python
import numpy as np
from sklearn.model_selection import train_test_split

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

```python
import pandas as pd
from google.colab import files
upload = files.upload()
```

```python
# Memasukkan dataset ke df
df = pd.read_csv('Covid Dataset.csv')

# Menampilkan dataset
df.head()
```

```python
# mengubah value yes dan no
df = df.replace({'Yes': 1, 'No': 0})

# Menampilkan kembali dataset
df.head()
```

```python
# Mengecek nilai null
df.isnull().sum()
```

```python
# Melihat informasi dataset
df.info()
```

```python
# Memasukkan columns dataset yang diperlukan kedalam X dan Y
X = df[['Breathing Problem','Fever','Dry Cough','Sore throat','Hyper Tension','Abroad travel','Contact with COVID Patient','Attended Large Gathering','Visited Public Exposed Places','Family working in Public Exposed Places']]
Y = df['COVID-19']
```

```python
# melihat isi X
print(X)
```

```python
# Melihat isi Y
print(Y)
```

```python
# Membagi dataset dengan perbandingan 80 : 20 ( train : test )
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Membangun model sequential
model = Sequential()

# Input layer dengan 10 unit neuron
model.add(Dense(units=10, activation='relu', input_dim=10))

# Hidden layer pertama dengan 10 unit neuron
model.add(Dense(units=10, activation='relu'))

# Hidden layer kedua dengan 6 unit neuron
model.add(Dense(units=6, activation='relu'))

# Output layer dengan 1 unit neuron (karena ini adalah masalah regresi)
model.add(Dense(units=1, activation='linear'))

# Mengompilasi model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Menampilkan ringkasan dari model
model.summary()
```

```python
# Memproses model
model.fit(X_train, Y_train,
                    epochs=5*87,
                    batch_size=32)
```

```python
# Evaluasi model pada data uji
loss, accuracy = model.evaluate(X_test, Y_test)

# Mencetak akurasi
print('Accuracy: %.2f%%' % (accuracy * 100))
```

Jadi kesimpulannya model yang dibuat menggunakan Algoritma ANN untuk mendeteksi apakah seseorang terkena covid atau tidak mendapatkan hasil yang cukup baik karena akurasinya 98,07% ketika diuji menggunakan data test
